```python
## Imports

# Display plots directly in the notebook instead of in a new window
%matplotlib inline

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
```

## Import the stats CSV into a DataFrame

Here we import the Stats DataFrame into the Phython notebook and explain its contents. This DataFrame is named `df_stats`.

   With the `ID` as the index. The `ID` refers to the Pokemon's unique Pokedex entry number. Each Pokemon has a unique Pokedex entry number and this number can be used to identify each of the Pokemon. 

   The `Pokemon` column is the name of the specifc Pokemon. This list includes all of the Pokemon up until Generation 7 of the main Pokemon series games. This does not include Meltan and Melmetal, Pokemon that were introduced outside of the main series games.

   The `Legendary` column of the DataFrame simply indicates whether or not the Pokemon is Legendary. If it is not Legendary, the `Legendary` column will be marked as False, and vice versa. 

   The `Stat Total` column reflects the aggregate stats of each of the Pokemon. This stat total is comprised of 6 different categories: Attack, Special Attack, Defense, Special Defense, HP and Speed. Every Pokemon has a base stat total for each of these 6 categories.

   The `ATK Sum` coulumn is the aggregate of each Pokemon's Attack Stats. We defined this as this being comprised of Attack, Special Attack, and Speed from the 6 possible stat categories. 

   The `DEF Sum` coulumn is the aggregate of each Pokemon's Defense Stats. We defined this as this being comprised of Defense, Special Defense, and HP from the 6 possible stat categories. 

```python
#Import DataFrame
df_stats= pd.DataFrame(pd.read_csv('statsDF.csv'))

#Set "ID" as the index
df_stats=df_stats.set_index('ID')
```

```python
#Display the DataFrame
df_stats
```

## Cleaning up the DataFrames

Because in this section we're only looking at the statistics of Pokemon in terms of their Attack, Defense and Stat Totals, we can drop columns that describe their physical characteristics (`Height` and `Weight`).


```python
#Drop unnecessary columns
df_stats=df_stats.drop(['Height'], axis=1)
df_stats=df_stats.drop(['Weight'], axis=1)

#Display the new DataFrame
df_stats
```

## Separating the DataFrames

Create two distinct DataFrames, one with only non-Legendary Pokemon and the other with only Legendary Pokemon. This just facilitates the cleaning process and makes looking at either Legendary or non-Legendary Pokemon easier. `df_legendary` is a DataFrame that holds the stats for only the Legendary Pokemon, and `df_normal` is a DataFrame that holds the stats of only the non-Legendary Pokemon.

```python
#Separate DataFrame into two DataFrames, Legendary and Non-Legendary
df_legendary= df_stats.loc[df_stats["Legendary"]==True]

df_normal=df_stats.loc[df_stats["Legendary"]==False]
```

Delete the "Legendary" column because we have separated the two and it's redundant.

```python
#Drop the unnecessary DataFrames
df_legendary= df_legendary.drop(["Legendary"], axis=1)

#Display the updated DataFrame
df_legendary
```

Create a new DataFrame that is just the averages of all the stat columns for the legendary Pokemon. We call this `df_legData`.

```python
#Average each of the columns of the df_legendary DataFrame
legStatTotal=df_legendary['Stat Total'].mean()
legAtkSum=df_legendary['ATK Sum'].mean()
legDefSum=df_legendary['DEF Sum'].mean()

#Create a DataFrame from the dictionary of this mean data
legData = {'Stat Total':[legStatTotal],'ATK Sum':[legAtkSum], "DEF Sum": [legDefSum]}
df_legData=pd.DataFrame.from_dict(legData)

#Display the new DataFrame
df_legData
```

Do exactly what you did for the Legendary Pokemon but for the non-Legendary Pokemon. Call the resulting DataFrame `df_normData`.

```python
#Drop the unnecessary column
df_normal= df_normal.drop(["Legendary"], axis=1)

#Display the updated DataFrame
df_normal
```

```python
#Average each of the columns of the df_normal DataFrame
normStatTotal=df_normal['Stat Total'].mean()
normAtkSum=df_normal['ATK Sum'].mean()
normDefSum=df_normal['DEF Sum'].mean()

#Create a new DataFrame that holds the averages of the df_normal
normData = {'Stat Total':[normStatTotal],'ATK Sum':[normAtkSum], "DEF Sum": [normDefSum]}
df_normData=pd.DataFrame.from_dict(normData)

#Display the new DataFrame
df_normData
```

Now that we have the averages, we can concatenate the two DataFrames into one. Call this `df_avgData`.

```python
#Create a DataFrame that is contains both 
df_avgData = pd.DataFrame({
    "Status":["Legendary","Normal"],
    "Stat Total": [legStatTotal, normStatTotal],
    "ATK Sum": [legAtkSum,normAtkSum],
    "DEF Sum": [legDefSum, normDefSum]})

#Display the DataFrame
df_avgData
```

## Visualizing the Data

**Radar Plot**


Here we create a Radar Plot that demonstrates the average stats all of the normal Pokemon and another that demonstrates the average stats of all the Legendary Pokemon. This will help illustrate the disparity that exists between the two stats of these different types of Pokemon. 

The following graph made by u/ibetDELWYN from reddit was used as inspiration to use this type of graph: 
https://i.imgur.com/3USj3xy.png

The area highlighted by the blue is the difference in stats between the legendary and non-legendary Pokemon. 

```python

# Define the number of variables
categories=list(df_avgData)[1:]
N = len(categories)
 
# Determine the angle of each section
angles = [n / float(N) * 2* pi for n in range(N)]
angles += angles[:1]
 
# Initialise the radar plot
ax = plt.subplot(111, polar=True)
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([100,200,300,400,500], ["100","200","300","400","500"], color="grey", size=7)
plt.ylim(0,650)

# Add Legendary Pokemon Stats to the graph
values=df_avgData.loc[0].drop('Status').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, 'orange', linewidth=1, linestyle='solid', label="Legendary")
ax.fill(angles, values, 'orange', alpha=0.1)

## Add non-Legendary Pokemon Stats to the graph
values=df_avgData.loc[1].drop('Status').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values,'b', linewidth=1, linestyle='solid', label="Normal")
ax.fill(angles, values, 'b', alpha=0.1)

## Add a Legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title("Radar Plot of Average Stats", pad=(20))

plt.show()
```

**Scatter Plot**

In this case we expect the legendary Pokemon to be on the extremes. Meaning that they have higher Defense stats, Attack stats, or a combination thereof. The combination of the two graphs supports this hypothesis, as the red dots are mainly congregated toward the top right of the graph, suggesting high defense and Attack.

## Predictions

The following code is the prediction for the stats of a new Legendary Pokemon when given either the `ATK Sum` or the `DEF Sum`. 

**Predict Total Stats from ATK Sum**

```python
#Create a polyfit to fit a 1-degree linear model 
a1,b1= np.polyfit(df_legendary['ATK Sum'],df_legendary['Stat Total'] , 1)
```

For demonstration purposes, we will predict the Defense Sum for a Legendary Pokemon that has 150 total attack stats.

```python
#Use polyfit to make a prediction 
new_atk=150

pred_total1= a1 * new_atk + b1

print("Predicted Total Stats for a Legendary Pokemon with 150 ATK Sum is", pred_total1)
```

**Predict Total Stats from DEF Sum**

This is the same process but instead of using `ATK Sum` we substitute it for `DEF Sum`.

```python
#Create a polyfit to fit a 1-degree linear model 
a2,b2= np.polyfit(df_legendary['DEF Sum'],df_legendary['Stat Total'] , 1)
```

This time, we will assume that the we know the total defense stats is 150 and we want to know the total stats.

```python
#Use polyfit to make a prediction 
new_def=150

pred_total2= a2 * new_def + b2

print("Predicted Total Stats for a Legendary Pokemon with 150 DEF Sum is", pred_total2)
```
