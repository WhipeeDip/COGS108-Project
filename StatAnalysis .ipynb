# B) Average Height contrasting Legendary & Non-Legendary Pokemons

```python
sns.boxplot(x=statsDF['Legendary'],y=statsDF['Height'])
plt.title("Comparison of Height between Legendaries & Non-Legendaries")
plt.xticks([False, True], ['Non-Legendary', 'Legendary'])
plt.xlabel("Pokemon")
```

# C) Average Weight contrasting Legendary & Non-Legendary Pokemons

```python
sns.boxplot(x=statsDF['Legendary'],y=statsDF['Weight'])
plt.title("Comparison of Weight between Legendaries & Non-Legendaries")
plt.xticks([False, True], ['Non-Legendary', 'Legendary'])
plt.xlabel("Pokemon")
```

Similar to the Height box plot, Legendary Pokemon on average tend to be heavier than Non-Legendaries. Again, there is a sizable amount of non-Legendary Pokemons that weigh above their average weight - varying from slightly to substantially heavier. The exception for this case is that there are also a good amount of Legendary Pokemon above their average weight, although not as much as Non-Legendaries.
